```python
# Outliers Identification
#There are different ways and methods of identifying outliers, but we are only going to use some of the most popular techniques:
#Visualization
#Skewness
#Interquartile Range
```

```python
sns.boxplot(data['EstimatedSalary'],data=data)
```

```python
Q1=data['EstimatedSalary'].quantile(0.25)
Q3=data['EstimatedSalary'].quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
#Outliers Treatment
# 1. Flooring and Capping.
# 2. Trimming.
```

```python
Q1=data['EstimatedSalary'].quantile(0.25)
Q3=data['EstimatedSalary'].quantile(0.75)
IQR=Q3-Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
data['EstimatedSalary']=np.where((data['EstimatedSalary'])>upper_whisker,upper_whisker,np.where(data['EstimatedSalary']<lower_whisker,lower_whisker,data['EstimatedSalary']))
```

```python
# perform encoding
# one Hot Encoding
# Label Encoding
```

```python
from sklearn.preprocessing import OneHotEncoder


Surname_encoder = OneHotEncoder()
Surname_reshaped = np.array(data_categorical['Surname']).reshape(-1, 1)
Surname_values = Surname_encoder.fit_transform(Surname_reshaped)

print(data_categorical['Surname'][:5])
print()
print(Surname_values.toarray()[:5])
print()
print(Surname_encoder.inverse_transform(Surname_values)[:5])
```

```python
Geography_encoder = OneHotEncoder()
Geography_reshaped = np.array(data_categorical['Geography']).reshape(-1, 1)
Geography_values = Geography_encoder.fit_transform(Geography_reshaped)

print(data_categorical['Geography'][:5])
print()
print(Geography_values.toarray()[:5])
print()
print(Geography_encoder.inverse_transform(Geography_values)[:5])
```

```python
Gender_encoder = OneHotEncoder()
Gender_reshaped = np.array(data_categorical['Gender']).reshape(-1, 1)
Gender_values = Gender_encoder.fit_transform(Gender_reshaped)

print(data_categorical['Gender'][:5])
print()
print(Gender_values.toarray()[:5])
print()
print(Gender_encoder.inverse_transform(Gender_values)[:5])
```

```python
from sklearn.preprocessing import LabelEncoder

Gender_encoder = LabelEncoder()
```

```python
Gender_encoder = LabelEncoder()
Gender_values = Gender_encoder.fit_transform(data_categorical['Gender'])

print("Before Encoding:", list(data_categorical['Gender'][:5]))
print("After Encoding:", Gender_values[:5])
print("The inverse from the encoding result:", Gender_encoder.inverse_transform(Gender_values[:5]))
```
